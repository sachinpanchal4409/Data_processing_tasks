```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
i=1
img1 = cv2.imread('C:/SachinP/Paper4_LASkipModel/DRIVE_Predicted_Out/groundtruth/DRIVE_groundtruth'+str(i)+'.png')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
h, w = img1.shape
#cv2.imshow('Hi', img1)
#cv2.waitKey(0)
#cv2.destroyAllWindows()

img2 = cv2.imread('C:/SachinP/Paper4_LASkipModel/DRIVE_Predicted_Out/Proposed_model/DRIVE_Pred_out_Porposed'+str(i)+'.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img3 = cv2.imread('C:/SachinP/Paper4_LASkipModel/DRIVE_Predicted_Out/AttUnet/DRIVE_Pred_out_AttUnet'+str(i)+'.png')
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)

def error(img1, img2):
   diff = cv2.subtract(img2, img1)
   err = np.sum(diff**2)
   mse = err/(float(h*w))
   msre = np.sqrt(mse)
   return msre, diff

match_error12, diff12 = error(img1, img2)
match_error13, diff13 = error(img1, img3)
match_error23, diff23 = error(img2, img3)

print("Image matching Error between image 1 and image 2:",match_error12)
print("Image matching Error between image 1 and image 3:",match_error13)
print("Image matching Error between image 2 and image 3:",match_error23)

plt.subplot(221), plt.imshow(diff12,'Blues_r'),plt.title("image1 - Image2"),plt.axis('off'),plt.savefig('diff_gt_pro.png',dpi=200)
plt.subplot(222), plt.imshow(diff13,'gray'),plt.title("image1 - Image3"),plt.axis('off'),plt.savefig('diff_gt_Unet.png',dpi=200)
plt.subplot(223), plt.imshow(diff23,'gray'),plt.title("image2 - Image3"),plt.axis('off'),plt.savefig('diff_pro_unet.png',dpi=200)
plt.show()
```
